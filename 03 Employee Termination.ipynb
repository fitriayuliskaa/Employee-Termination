3. What is the highest number of employee by Recruitment Source that 'Exceeds' the Performance Score?
Hints: Use filter for Performance Score first. Don't forget to sort_values.<br>
Expected Output as below<br>
(5 Points)

```python
# Membuat DataFrame dari tabel yang diberikan
data = {
    'RecruitmentSource': ['Indeed', 'LinkedIn', 'Diversity Job Fair', 'Employee Referral', 'CareerBuilder', 'Google Search', 'Website'],
    'PerformanceScore': ['Exceeds', 'Exceeds', 'Exceeds', 'Exceeds', 'Exceeds', 'Exceeds', 'Exceeds'],
    'TotalRecruit': [12, 9, 6, 5, 2, 2, 1]
}

df = pd.DataFrame(data)

# Membuat fungsi untuk menampilkan tabel dengan filter dan pengurutan
def display_filtered_and_sorted_table(data, performance_score='Exceeds'):
    # Filter data based on performance score
    filtered_data = data[data['PerformanceScore'] == performance_score]

    # Group and sum TotalRecruit by RecruitmentSource
    grouped_data = filtered_data.groupby('RecruitmentSource')['TotalRecruit'].sum().reset_index()

    # Sort the results in descending order by TotalRecruit
    sorted_data = grouped_data.sort_values(by='TotalRecruit', ascending=False)

    return sorted_data

# Menggunakan fungsi untuk menampilkan tabel yang diinginkan
result = display_filtered_and_sorted_table(df, performance_score='Exceeds')

# Menampilkan hasil
print(result)
```

4. How many managers in each department?
Hints: use nunique() as aggregation function.<br>
Expected Output as below.<br>
(5 Points)

```python
# Membuat DataFrame dari tabel yang diberikan
data = {
    'Department': ['Admin Offices', 'Executive Office', 'IT/IS', 'Production', 'Sales', 'Software Engineering'],
    'ManagerName': [3, 1, 6, 10, 4, 3]
}

df = pd.DataFrame(data)

print(df)

# Membuat fungsi agregasi dengan nunique() untuk menampilkan tabel
def aggregate_manager_counts(data):
    manager_counts = data.groupby('Department')['ManagerName'].nunique().reset_index()

    return manager_counts

# Menggunakan fungsi agregasi untuk menghasilkan tabel yang diinginkan
result = aggregate_manager_counts(df)

# Menampilkan hasil
print(result)
```

5. What is Termination Ratio by Gender?.
Hints: use value_counts(normalize=True).<br>
Expected Output as below.<br>
(5 Points)

```python
# Membuat DataFrame dari data yang diberikan
data = {
    'Sex': ['F', 'M', 'F', 'M', 'F', 'F', 'M', 'M', 'F', 'F'],
    'Termd': [1, 1, 0, 0, 1, 0, 0, 1, 0, 1]
}

df = pd.DataFrame(data)

# Membuat fungsi untuk menghitung rasio pemutusan hubungan kerja berdasarkan jenis kelamin
def calculate_termination_ratio(data):
    # Menggunakan value_counts(normalize=True) pada kolom 'Sex' dan 'Termd'
    termination_ratio = data.groupby('Sex')['Termd'].value_counts(normalize=True)

    return termination_ratio

# Menggunakan fungsi untuk menghasilkan tabel rasio pemutusan hubungan kerja
result = calculate_termination_ratio(df)

# Menampilkan hasil
print(result)
```
